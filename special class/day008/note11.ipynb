# 뉴런의 출력이 하나인 이유
여러개의 output이 나오면 더 복잡하고 성능이 좋지 않을까???  
여러 다른 값의 출력이 나오는 뉴런을 하나의 출력을 하는 뉴런으로 대체 가능해서  
딱히 여러개의 출력일 필요가 없다.  

output이 다른건 가중치 때문이다.  

네트워크가 아무리 복잡해도 노드의 결과는 선형방정식이다.  
현실 문제에서는 비선형이 많아서, 비선형적인 요소를 넣어야 한다. (activation 함수)  

ex) n_sample =1000, batch_size = 100 라면,

mini-batch 안쓰면, 한 epoch에 1번(1000/1000) minimum을 향해서 이동  
mini-batch 사용하면, 한 epoch에 10번(1000/100) minimum을 향해서 이동

원래 실제로는 batch_size는 전체 데이터 수와 같아야 한다. 정확하긴 하지만 느리다.    
mini batch 방법은 부정확하지만 더 많이 움직이는게 학습이 더 효율적으로 일어남  

gradient == 가장 크게 증가하는 방향이므로 -1을 곱해서 가장 작아지는 방향으로 향하게 한다.
