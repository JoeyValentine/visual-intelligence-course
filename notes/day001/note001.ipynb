```
# constant : 학습을 해도 변하지 않는 값
```

```
W = tf.Variable(tf.random_normal(shape=[1], name='w')) # 학습 대상
b = tf.Variable(tf.random_normal(shape=[1], name='b')) # 학습 대상
x = tf.placeholder(tf.float32) # 입력 값

pred = W * x + b  # 모델
```

```
y = tf.placeholder(tf.float32)  # 정답
loss = tf.reduce_mean(tf.square(pred - y))  # 정답에서 내가 구한 값을 뺀 값의 
# 제곱의 평균을 손실함수로
# reduce : 결과를 구할 때 '차원을 줄인다'라는 의미에서 reduce
```

```
# optimizer 손실값을 구하고 W와 b를 어떻게 수정할까 결정함
optimizer = tf.train.GradientDescentOptimizer(0.01)  
# 0.01 learning rate 
# 너무 느리면 최적해를 못 찾는다.
# 너무 크면 : 학습이 stable하게 안 됨 
train_step = optimizer.minimize(loss)  # 한 번의 훈련 과정이라 정의
```

```
# y = 2x + 1
x_train = [1,2,3,4]
y_train = [3,5,7,9]
```

```
sess = tf.Session()
result = sess.run(tf.global_variables_initializer())  
# W와 b의 입력값을 랜덤으로 초기화해줌
```

```
for i in range(1000):
  sess.run(train_step, feed_dict={x: x_train, y: y_train})
# x에 x_train를 먹여(feed)준다...!
```

```
x_test = [3, 5, 5 ,6]

print(sess.run(pred, feed_dict={x: x_test}))
```
